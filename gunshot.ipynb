# A simple model of gunshot in the air

The recent news about gunshot in to the air may kill the others. We construct a calculation to find the horizontal and vertical distance in vary angle, neglect the air resistance and the change of air density in high altitude. Then plot the graph of bullet's trajectory.

Our bullet here is 9Ã—19mm Parabellum (https://en.wikipedia.org/wiki/9%C3%9719mm_Parabellum) which has the mass 7.45-8.04 g depend on type. Here we use 7.45 g (115 gr) velocity 360 m/s Federal FMJ for a test. Only the velocity is use in this calculation.

### Create a function calculate horizontal and vertical distance and plot the graph.

Import library.

```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

Define the calculate function.

```python
def cal_distance(u, thetas):
    for theta in thetas:
        #save the input in degrees format
        degrees = theta
        #math library accepct theta in radian, we must change it first.
        theta = theta*math.pi/180

        #find the cos and sin of start velocity.
        u_x = u*math.cos(theta)
        u_y = u*math.sin(theta)

        #finding the time to reach maximum height.
        t_ver_max = u_y/9.81
        #finding the maximum time (the bullet hit the ground).
        t_max = t_ver_max*2
        #vertical distance
        s_y_max = (u_y*t_ver_max) + (-9.81*0.5*math.pow(t_ver_max,2))
        #if there is a very small value, show as 0.
        if s_y_max < 0.00001:
            s_y_max = 0
        #horizontal distance
        s_x_max = (u_x*t_max)
        #if there is a very small value, show as 0.
        if s_x_max < 0.00001:
            s_x_max = 0

        #Create series of time to use in for loop.
        time = np.arange(0, t_ver_max, 0.1)
        #Initial the list s_x and s_y use to collect the coordinate x and y.
        s_x = []
        s_y = []
        for t in time:
            x = u_x*t
            y = u_y*t + (-9.81*math.pow(t, 2))
            s_x.append(x)
            s_y.append(y)

        #plot graph.
        plt.xlabel('horizontal distance')
        plt.ylabel('vertical distance')
        plt.plot(s_x, s_y, label=str(degrees) + " degrees")
        
        #print the detail of each theta.
        print("Maximum horizontal distance " + str(degrees) + " degrees:", s_x_max)
        print("Maximum vertical distance " + str(degrees) + " degrees:", s_y_max)
        print("Maximum time before bullet hit the ground: ", t_max)
        print('\n')
    
    plt.legend()
    plt.show()
```

```python
cal_distance(360, [15, 30, 45, 75, 90])
```
